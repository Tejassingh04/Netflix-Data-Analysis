Netflix Data Analysis


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('netflix_titles.csv')
df.head()

top_genres = df['listed_in'].str.split(',', expand=True).stack().value_counts().head(3)

type_count = df['type'].value_counts()

top_countries = df['country'].dropna().str.split(', ', expand=True).stack().value_counts().head(5)

year_trends = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(6,4))
type_count.plot(kind='bar')
plt.title("Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```
